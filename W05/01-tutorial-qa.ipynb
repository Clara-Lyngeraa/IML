```python
# Write your solution here
# 2: If a=(1, 0,7) then the x_prime is closer to being x, but it is impossible to get x_prime = x
# 5: The projection matrix is idempodent, meaning P*P=P.
# importing libraries
import numpy as np
import matplotlib.pyplot as plt

# Three points
X = np.array([
    [1, 2],
    [2, 1.5],
    [3, 1.2]
]).T

# Show plot
plt.scatter(X[0, :], X[1, :], c="r")

# Make line points (remember Numpy broadcasting)
x = np.linspace(0, 4)
f_x = x * 0.5

# Plot line
# Add grid lines
plt.grid(True)
plt.plot(x, f_x)
plt.xlabel("X coordinates")
plt.ylabel("Y coordinates")
plt.show()

##1
#The line l written as the design matrix
A = np.array([[1, 0.5]]).T  #  has to be a column vector

##2
## construct projection matrix
P = (A @ np.linalg.inv(A.T @ A)) @ A.T
print("P:\n", P)

#projection the points with matrix multiplication
x_prime = P @ P @ X
print("projected points:\n", x_prime)
print(" X:\n", X)
print(" p * p:\n", P @ P )

```
